### IRIS dataset

- 데이터 분석 실습에서 가장 많이 활용되는 붓꽃 데이터셋
- 총 3개의 붓꽃 종에 대한 꽃잎과 꽃밭침의 길이 및 너비 측정 데이터가 각 종마다 50개씩 총 150개 저장

```python
# 라이브러리
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn import datasets   # iris 데이터 셋
```

```python
# 한글 폰트 지정
font_name = matplotlib.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
matplotlib.rc('font', family = font_name)
```

```python
# 데이터셋 가져오기
dataset = datasets.load_iris()
dataset
```

```python
# 150개의 꽃잎, 꽃받침 수치 데이터
dataset['data']
```

```python
# 데이터 상세 분류(꽃잎 너비, 꽃잎 길이, 꽃받침 너비, 꽃받침 길이)
dataset['feature_names']
```

```python
# 3개의 종 이름
dataset['target_names']
```

#### 1. 꽃잎, 꽃받침 수치를 데이터로, 속성명을 컬럼으로 가지는 데이터프레임 생성

```python
df = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
df.head()
```

#### 2. target 데이터와 target_names 데이터를 결합하여 종이름 데이터를 컬럼으로 추가

```python
# 각 수치형 target을 문자형/명목형 데이터로 변형하여 컬럼으로 결합
# 0 -> setosa, 1 -> versicolor, 2 -> virginica
# [(k, v)]
# zip([1, 2, 3], ['a', 'b', 'c'])
species_map = dict(zip([0, 1, 2], dataset['target_names']))
species_map
```

```python
# 새로운 컬럼인 species 생성
# 컬럼 인덱싱 = 리스트
df['species'] = [species_map[i] for i in dataset['target']]
[species_map[i] for i in dataset['target']]
```

```python
# 최종 데이터 확인
df
```

### Pie Plot(원형 차트)
- 전체에서 특정 범주가 차지하는 비중/비율을 파악하기 위해 사용
- pandas.DataFrame/Series.plot(kind='pie')
    - Series 객체 : 하나의 원형 차트 생성
    - DataFrame 객체 : 각 열 마다 하나씩 원형 차트 생성
- pandas.DataFrame/Series.plot.pie(y=None, figsize=None, subplots=False)
    - y : 하나의 컬럼에 대해 원형 차트를 생성할 때 사용할 컬럼명
    - figsize : 생성할 figure 객체 크기(tuple)
    - subplots : y 값으로 하나의 컬럼을 사용하는 것이 아니라 전체 df를 사용할 때 True로 지정

```python
# 전체 종별 데이터 평균
df.head()
# groupby 혹은 pivot_table을 이용해서 종별 데이터 평균을 구해주세요.
mean_by_species = df.groupby('species').mean()
mean_by_species

mean_by_species2 = df.pivot_table(index='species', values=['sepal length (cm)',
                        'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'],
                                 aggfunc='mean')
mean_by_species2
```

```python
# 하나의 속성(feature)별로 종별 파이 차트 생성
# 속성별로 종별 비중
# subplots : 컬럼별로 subplot 생성 여부
# layout : 튜플 타입으로 행, 열, 구조 지정
# legend : 범례 여부
mean_by_species.plot(kind='pie', subplots=True, layout=(2, 2), legend=False)
```

```python
mean_by_species.plot(kind='pie', y='sepal length (cm)', legend=False,
                    label='', title='< sepal length >')
# label : 사용한 컬럼명 표기 여부
# title : 차트 자체의 제목 지정
```

### Histogram(히스토그램)

- pandas.DataFrame/Series.plot(kind='hist')
    - Series 객체 : 하나의 히스토그램 생성
    - DataFrame 객체 : 각 열 마다 하나씩 히스토그램 차트 생성,<br> subplots=True & layout = (n,m)
- pandas.DataFrame/Series.hist(bins=10)
    - bins : 히스토그램 구간 수를 나타내는 값으로 구간 수가 너무 작으면(=간격이 너무 넓으면) 세부 분포 파악이 어렵고, 구간 수가 너무 많으면 전체적인 분포 파악이 어렵다.

```python
# 두 번째 방법
df.hist()
```

```python
df.hist(bins=3)
```

```python
# 종 별로 petal length에 대한 히스토그램
# ver1. df.plot(kind='hist')
# 종별로 데이터프레임 구분 + plot
for spec in df['species'].unique():
    tmp = df[df['species'] == spec]
    tmp['petal length (cm)'].plot(kind='hist')
    
plt.legend()
plt.suptitle('종별 꽃잎 길이')
plt.show()
```

```python
# ver2. plot(kind='hist') - ver1의 문제 해결
for spec in df['species'].unique():
    tmp = df[df['species']==spec]
    # label -> legend에서 보여질 이름, grid -> 격자선 표시 여부
    tmp['petal length (cm)'].plot(kind='hist', label=spec, grid=True)
    
plt.legend()
plt.suptitle('종별 꽃잎 길이')
plt.show()
```

### scatter-plot(산포도)

- 두 종류의 변수만 포함하여 적용할 수 있음
- 두 변수 간의 관계에 대해 분산 분포를 통해 파악할 수 있음
- pandas.DataFrame/Series.plot(kind='scatter', x=None, y=None)
    - x : 가로축(x축)에 사용될 데이터/컬럼명
    - y : 세로축(y축)에 사용될 데이터/컬럼명
- pandas.DataFrame/Series.plot.scatter(x, y, s=None, c=None)
    - x : 가로축(x축)에 사용될 데이터/컬럼명
    - y : 세로축(y축)에 사용될 데이터/컬럼명
    - s : 각 점의 크기
    - c : 각 점의 색상

```python
# 모든 종에 대하여 꽃받침 길이(petal length) 대 너비(petal width) 데이터 분산 분포
df.drop('idx', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.plot(kind='scatter', x='petal length (cm)', y='petal width (cm)')
plt.title('< 꽃받침 길이 대 너비의 관계 >')
plt.show()
```

```python
# 종 기준으로 산포도 색상을 다르게 지정
# colormap 사용을 위해 구분값을 숫자로 가지는 컬럼을 추가
# 범주형 데이터 -> 숫자형 데이터로 매핑(마커 색깔을 숫자로 지정 가능하기 때문에...)
df['species_map'] = dataset.target
df.head()
```

```python
# 산점도 색상이 요소별로 다르게 지정된 경우
df.plot.scatter(x='petal length (cm)', y='petal width (cm)',
               c='species_map', colormap='plasma')
```
